to load a google sheet into jupyter via API: 
https://towardsdatascience.com/from-google-sheet-to-your-jupyter-notebook-ccdbf28fbf1b

- Go to https://console.developers.google.com/project and create a new project (name doesn't matter)
- Navigate to the project -> APIs&Services -> Dashboard
- Click "Enable APIs," find Google Sheets API by searching for it, and enable it
- You'll be taken to the Google Sheets API dashboard
- Click "Create credentials" and choose "Service account"
- Give the account a name (name doesn't matter) and for role, select "Service account user" by searching for it
- Navigate to the "Key" tab to create a key, download it in .json format and save to the directory of your Jupyter notebook
- Open the .json and copy the client email. Go to the Google Sheet you want to use and click "Share" and paste in the service account email
- Open a new Jupyter notebook and paste the proper info as shown in the following 3 cells

you only have to do the steps once (per google sheet), and then open a notebook and load in the sheet and it will update to reflect all changes in the google sheet since the previous time 

```python
# import dependencies 

import pandas as pd 
import gspread 
from oauth2client.service_account import ServiceAccountCredentials
```

```python
# set up credentials 

scope = ['https://spreadsheets.google.com/feeds'] 

credentials = ServiceAccountCredentials.from_json_keyfile_name('spartan-network-323600-a7c56fe521c4.json', scope)
    #for the .json file name you'll need to use the name of the file containing your service account key
    
gc = gspread.authorize(credentials)
```

```python
#to save current version as a csv

import csv

date='08-23'

with open('ChemicalInventory_{}.csv'.format(date), 'w', newline='') as file:
    writer = csv.writer(file)
    for line in table:
        writer.writerow(line)
```
